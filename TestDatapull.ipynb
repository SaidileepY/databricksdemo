```python
# Read orders file from ABFS bronze container using Auto Loader
df = (spark.readStream
  .format("cloudFiles")
  .option("cloudFiles.format", "parquet")  # Change to json, csv, etc. as needed
  .option("cloudFiles.schemaLocation", "/tmp/schema/orders")  # Schema inference location
  .load("abfss://bronze@<storage-account>.dfs.core.windows.net/orders"))

display(df)
```

```python
# Write streaming data to silver layer in Delta format
(df.writeStream
  .format("delta")
  .option("checkpointLocation", "/tmp/checkpoints/orders_silver")
  .outputMode("append")
  .start("abfss://silver@<storage-account>.dfs.core.windows.net/orders"))
```

```python
# Read finance file from ADLS silver container in batch mode
df_finance = spark.read.format("parquet").load("abfss://silver@<storage-account>.dfs.core.windows.net/finance")

display(df_finance)
```
